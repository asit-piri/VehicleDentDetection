```python
import json
import os
```

```python
path = '/mnt/01D557900A25E360/Study/unet-research/output_restructured.json'
dent_annotation = json.load(open(path))
```

```python
file_json_path = '/mnt/01D557900A25E360/Study/unet-research/file_json'
```

```python
id_name = {}
for ids in dent_annotation['images']:
    id_name[ids.get('id')] = ids.get('file_name') 
```

```python
id_name
```

```python
for annt in dent_annotation.get('annotations'):
#     try:
    fn = id_name.get(annt.get('image_id')).split('.')[0]
    print(fn , annt.get('image_id'))
    json_object = json.dumps(annt, indent = 4) 
    with open(os.path.join(file_json_path, f'{fn}.json'), "w") as outfile: 
        outfile.write(json_object)
#     except Exception as ex:
#         print(f'{annt.get("id")} is not available')
```
