# Using Unet package

## Necessary Imports

```python
from unet.UNet import UNet
```

## Set the Parameters

```python
params = {
    'loss': 'sparse_categorical_crossentropy',
    'batch_size': 16,
    'optimizer': 'sgd',
    'epochs': 2,
    'image_shape': (224, 224),
    'seed': 47,
    'apply_augmentation': True,
    'augmentation_threshold': 0.4,
    'checkpoint_path': '.',
    'train_path': ('data/train_images', 'data/train_pixel_masks'),
    'val_path': ('data/test_images', 'data/test_pixel_masks'),
    'test_path': ('data/test_images', 'data/test_pixel_masks')
}
```

## Generate Train and Test Datasets

```python
model = UNet(params)
```

```python
model.get_parameters()
```

```python
model.visualize_data()
```

## Train the Model

```python
model.train()
```

## Save the Model Config

This is for logging different train conditions

```python
model.save_model()
```

## Test the Model

```python
model.show_prediction_result()
```
